### <center>Extracción de datos</center>

Se observa que el archivo es de tipo JSON, por lo cual se elige una forma particular para poder abrirlo. Se extraen los datos a un DataFrame.

```python
#Se pasa dirección del archivo json
archivo = os.path.join('..', 'BD', 'output_steam_games.json')

# Se lee cada línea del dataset
registros= []
with open(archivo) as f:
    for line in f.readlines():
        data = json.loads(line)
        registros.append(data)

# Se transfoma en DataFrame
df_game = pd.DataFrame(registros)
```

Se procede a eliminar las columnas consideradas no necesarias para el estudio a realizar

```python
# Se elimina las columnas que no necesitamos del Dataframe
df_game = df_game.drop('publisher', axis=1)
df_game = df_game.drop('url', axis=1)
df_game = df_game.drop('app_name', axis=1)
df_game = df_game.drop('specs', axis=1)
df_game = df_game.drop('early_access', axis=1)
df_game = df_game.drop('tags', axis=1)
```

```python
# Se contabiliza la aparición de cada valor que toma la columna 'price'
conteo_publisher = df_game['price'].value_counts()
conteo_publisher
```

```python
# Se reemplaza 'Free To Play' por 0 y se convierte la columna 'price' a tipo float
df_game['price'] = pd.to_numeric(df_game['price'].replace('Free To Play', 0), errors='coerce')

# Reemplazar NaN por 0
df_game['price'] = df_game['price'].fillna(0)
```
