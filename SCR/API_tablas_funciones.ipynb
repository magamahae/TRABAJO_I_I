# <center> CONSULTA II
### <i><b>def UserForGenre( genero : str )</i></b>: </br>
 #### Debe devolver el usuario que acumula más horas jugadas para el género dado y una lista de la acumulación de horas jugadas por año..</br>
#### - Ejemplo de retorno: {"Usuario con más horas jugadas para Género X" : us213ndjss09sdf, "Horas jugadas":[{Año: 2013, Horas: 203}, {Año: 2012, Horas: 100}, {Año: 2011, Horas: 23}]}

_________________________

#### Carga de archivo PARQUET

```python
# Se obtiene la ruta completa al archivo PARQUET
csv_path = os.path.join('..', 'DATA', 'genre_use_year_2.parquet')

# Se lee el archivo 
df_2 = pd.read_parquet(csv_path)
```

____________________

#### Se visualizan los datos a cargar en la API

```python
# Se visualiza el dataframe
df_2
```

_________________

### Se genera la función

```python
def UserForGenre(genero):
    # Se filtra por el género especificado
    df_genre = df_2[df_2['genres'] == genero]

    # Si no hay datos para el género especificado, se retorna un mensaje
    if df_genre.empty:
        return f"No hay datos para el género '{genero}'"

    # Se agrupa por usuario y género y se calculan las horas jugadas totales
    df_genre_g = df_genre.groupby(['user_id'])['playtime_forever'].sum()

    # Se encontra el usuario con más horas jugadas
    max_playtime = df_genre_g.idxmax()

    # Se filtra por el usuario con más horas jugadas
    #df_user_max_playtime = df_genre[df_genre['user_id'] == max_playtime]

    # Se agrupa por año y se calculan las horas jugadas totales
    grouped_by_year = df_genre.groupby('release_year')['playtime_forever'].sum()

    # Se crea lista de acumulación de horas jugadas por año
    acumulacion_horas = [{'Año': year, 'Horas': hours} for year, hours in grouped_by_year.items()]

    # Se retornar el resultado como un diccionario
    return {"Usuario con más horas jugadas para Género {}".format(genero): max_playtime, "Horas jugadas": acumulacion_horas}
```

# <center> CONSULTA V
### <i><b>def sentiment_analysis( empresa desarrolladora : str )</i></b>: </br>
 ####  Según la empresa desarrolladora, se devuelve un diccionario con el nombre de la desarrolladora como llave y una lista con la cantidad total de registros de reseñas de usuarios que se encuentren categorizados con un análisis de sentimiento como valor.</br>
#### - Ejemplo de retorno: {'Valve' : [Negative = 182, Neutral = 120, Positive = 278]}

___________________________

#### Carga de archivo PARQUET

```python
# Se obtiene la ruta completa al archivo PARQUET
csv_path_as = os.path.join('..', 'DATA', 'df_developer_sanalysis.parquet')

# Se lee el archivo
df_as = pd.read_parquet(csv_path_as)
```

____________________

#### Visualización del dataframe a utilizar en la API


```python
# Se muestra el dataframe
df_as
```

__________________________

### Se genera la función

```python
def sentiment_analysis( empresa_desarrolladora : str): 
    # Filtrar el DataFrame por la empresa desarrolladora proporcionada
    developer_df = df_as[df_as['developer'] == empresa_desarrolladora]
    
    # Crear el diccionario de retorno
    result = {empresa_desarrolladora: {'Negative': 0, 'Neutral': 0, 'Positive': 0}}

    # Llenar el diccionario con la cantidad de registros para cada categoría de sentimiento
    for sentiment, count in zip(developer_df['sentiment_analysis'], developer_df['recommend_count']):
        sentiment_mapping = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
        sentiment_category = sentiment_mapping[sentiment]
        result[empresa_desarrolladora][sentiment_category] += count

    return result
  
```

_____________________________________________

### Se prueba la función

```python
sentiment_analysis('11 bit studios')
```

### <i><b>def recomendacion_juego( id de producto )</i></b>: </br>
 ####  Ingresando el id de producto, deberíamos recibir una lista con 5 juegos recomendados similares al ingresado.

___________________________

#### Carga de archivo PARQUET

```python
# Se obtiene la ruta completa al archivo PARQUET
csv_path_as = os.path.join('..', 'DATA', 'modelo_rec.parquet')

# Se lee el archivo
df = pd.read_parquet(csv_path_as)
```

_____________________________

### Se genera la función

```python
from fastapi.responses import JSONResponse


def recomendacion_juego(id_producto: int):
    recomendaciones = df[df['item_id'] == id_producto]['recomendaciones'].iloc[0]
    
    # Verificar si la lista de recomendaciones no está vacía
    if len(recomendaciones) > 0:
        recomendaciones_dict = {i + 1: juego for i, juego in enumerate(recomendaciones)}
        return recomendaciones_dict
    else:
        # Si no se encontraron recomendaciones para el ID, devolver un mensaje de error
        error_data = {'error': 'No se encontraron recomendaciones para el ID proporcionado'}
        return JSONResponse(content=error_data, status_code=404)
```

_________________________________________________________

### Respuesta obtenida

```python
recomendacion_juego(449940)
```
